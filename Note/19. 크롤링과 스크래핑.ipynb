# 데이터 다운로드 하기 -- 02

```python
import urllib.request

# URL과 저장경로 지정하기.
url = "http://uta.pw/shodou/img/28/214.png"
savename = "../Data/test02.png"

# 다운로드 하기
mem = urllib.request.urlopen(url).read()

# 파일로 저장하기
with open(savename, "wb") as f:
    f.write(mem)
    
print("저장되었습니다")
```

# 매개변수를 추가해 요청을 전송하는 방법
기창청 RSS : http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp   
URL 뒤에 매개변수를 입력하여 지역별로 일기 정보를 가져올 수 있음.     
전국 :  108, 서울/경기도 : 109 , 강원도 : 105, 제주특별자치도 : 184   
ex) 서울/경기도 : http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stdId=109

```python
import urllib.request
import urllib.parse

API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

# 매개변수를 url encoding
values = {'stnId':'109'}
params = urllib.parse.urlencode(values)

# 요청 전용 url생성
url = API + "?" + params
print("url = ", url)

data =  urllib.request.urlopen(url).read()
# print(data) # 글자를 확인할 수 없음
text = data.decode('utf-8')
print(text)
```

# beatifulsoup 로 scraping 하기
- BeautifulSoup : 간단하게 HTML과 XML에서 정보 추출
- Beautiful 설치: pip install beautifulsoup4

```python
# 기본 사용법
from bs4 import BeautifulSoup 

#HTML Sample
html = """
        <html>
            <body>
                <h1> Header </h1>
                <p> Line 1을 서술 </p>
            </body>
        </html>
"""

# Html 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
h1 = soup.html.body.h1
p = soup.html.body.p
print(h1)
print(p)

# text만 출력
print("h1 = ", h1.string)
print("p = ", p.string)
```

```python
# id로 요소를 추출하기

from bs4 import BeautifulSoup 

#HTML Sample
html = """
        <html>
            <body>
                <h1 id="title"> Header </h1>
                <p id="body"> Line 1을 서술 </p>
            </body>
        </html>
"""

# Html 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
title = soup.find(id='title')
body =  soup.find(id='body')
print(title)
print(body)

# text만 출력
print('title=', title.string)
print('body= ',body.string)

print('title =' , title.text) # string에서 안되면 text를 쓴다 
```

```python
# 여러개의 요소 추출하기
from bs4 import BeautifulSoup 

#HTML Sample
html = """
        <html>
            <body>
                <ul>
                    <li><a href="http://www.naver.com">naver</a></li>
                    <li><a href="http://www.daum.net">daum</a></li>
                </ul>
            </body>
        </html>
"""

# Html 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
links = soup.find_all('a')
print(links)

for link in links:
    href = link.attrs['href']
    text = link.string
    print(text,"==>", href)
```

```python
# 기상청 자료 활용하기
from bs4 import BeautifulSoup 
import urllib.request as req

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

# data 가져오기
res = req.urlopen(url)


# BeautifulSoup으로 분석
soup = BeautifulSoup(res, 'html.parser')
#print(soup)

# 원하는 데이터 추출하기
title = soup.find('title').string
wf = soup.find('wf').string

print(title)
print(wf)
```

```python
# 데이터 정리하기
list_wfs = list(wf)
except_char = ['<','b','r','/','>']
result = ''

for lwf in list_wfs:
    if lwf == '○':
        result +="\n"+lwf
    elif lwf not in except_char:
        result += lwf
 
        
print(title)
print("="*50)
print(result)
```

# css 선택자 사용하기
- soup.select_one() : css 선택자로 요소 하나를 추출
- soup.select() : css 선택자로 여러개를 리스트로 추출

```python
from bs4 import BeautifulSoup 

#HTML Sample
html = """
        <html>
            <body>
               <div id="meigen">
                    <h1>위키북스 도서</h1>
                  <ul class="items">
                    <li>유니티 게임 이펙트 입문</li>
                    <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
                    <li>모던 웹사이트 디자인의 정석</li>
                  </ul>
                </div>
            </body>
        </html>
"""

#HTML 분석하기
soup = BeautifulSoup(html,'html.parser')
      
#필요한 부분을 CSS 쿼리로 추출하기
#타이틀 부분 추출하기
h1 = soup.select_one("div#meigen > h1").string # id:#, class:. , tag : tag명 , 자손:> , 후손: ' '
print("h1 : ",h1)
      
#목록부분 추출하기
li_list = soup.select("div#meigen > ul.items > li")
for li in li_list:
    print("li : ",li.string)
```

### 윤동주 시인 작품 가져오기
https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC

```python
from bs4 import BeautifulSoup 
import urllib.request as req

# HTML 가져오기
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")

# 데이터 추출하기
a_list = soup.select("#mw-content-text > div.mw-parser-output > ul > li a")
# print(a_list)

for a in a_list:
    name = a.string
    if name == "증보판":
        continue
    print(name)
```

# Rotten Tomatoes   
www.rottentomatoes.com/top/bestofrt/?year=2019

```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = req.urlopen(url)
soup = BeautifulSoup(res, "html.parser")

```

### 테스트

```python
#top_movies_main > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > a
soup.select("#top_movies_main > div > table > tr > td > a")[0].string
```

## 적용

```python
li_lists = soup.select("#top_movies_main > div > table > tr > td > a")
count = 0
titleLists = []
countLists = []

for li in li_lists:
    count += 1
    print(count, li.string.strip())
    countLists.append(count)
    titleLists.append(li.string.strip())
```

```python
# Lists 확인
print(countLists)
print(titleLists)
```

```python
# countList만으로 dataframe 생성
df = pd.DataFrame(data=countLists, columns=['index'])
df
```

```python
# 기존 dataframe에 titleLists 추가
df['title'] = pd.Series(titleLists)
df
```

```python
# csv로 저장
df.to_csv("../Data/rottenTomatoesList_2019.csv", encoding='utf-8', index=False)
```
